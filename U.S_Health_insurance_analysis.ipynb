# **Univariate Analysis**

Univariate analysis focuses on analyzing each feature in the dataset independently.

- **Distribution analysis**: The distribution of each feature is examined to identify its shape, central tendency, and dispersion.

- **Identifying potential issues**: Univariate analysis helps in identifying potential problems with the data such as outliers, skewness, and missing values

```python
df.head()
```

## Steps of doing Univariate Analysis on Numerical columns

- **Descriptive Statistics**: Compute basic summary statistics for the column, such as mean, median, mode, standard deviation, range, and quartiles. These statistics give a general understanding of the distribution of the data and can help identify skewness or outliers.

- **Visualizations**: Create visualizations to explore the distribution of the data. Some common visualizations for numerical data include histograms, box plots, and density plots. These visualizations provide a visual representation of the distribution of the data and can help identify skewness an outliers.

- **Identifying Outliers**: Identify and examine any outliers in the data. Outliers can be identified using visualizations. It is important to determine whether the outliers are due to measurement errors, data entry errors, or legitimate differences in the data, and to decide whether to include or exclude them from the analysis.

- **Skewness**: Check for skewness in the data and consider transforming the data or using robust statistical methods that are less sensitive to skewness, if necessary.

- **Conclusion**: Summarize the findings of the EDA and make decisions about how to proceed with further analysis.

```python
df.describe()
```

```python
# Plotting a histogram of the numerical (continuous) columns

plt.figure(figsize=(13,5))

plt.subplot(1,3,1)
df['age'].plot(kind='hist',bins=10)
plt.xlabel('Age Values')
plt.title('Age Histogram')

plt.subplot(1,3,2)
df['bmi'].plot(kind='hist',bins=10,color='brown')
plt.xlabel('BMI Values')
plt.title('BMI Histogram')

plt.subplot(1,3,3)
df['charges'].plot(kind='hist',color='g')
plt.xticks(rotation=70)
plt.xlabel('Charges Values')
plt.title('Charges Histogram')

plt.suptitle('Histograms of all the Numerical Columns')

plt.tight_layout()
plt.show()
```

```python
# Plotting KDE-plots of the numerical (continuous) columns to visualize the distributions

plt.figure(figsize=(13,4))

plt.subplot(1,3,1)
sns.kdeplot(df['age'])
plt.title('Distribution of the "Ages"')

plt.subplot(1,3,2)
sns.kdeplot(df['bmi'])
plt.title('Distribution of the "BMIs"')

plt.subplot(1,3,3)
sns.kdeplot(df['charges'])
plt.title('Distribution of the "Charges')

plt.suptitle('Distributions of all the Numerical Columns')

plt.tight_layout()
plt.show()

print()

print('_'*185)

print()
print(f'Skewness of "age": {df["age"].skew():.2f}')
print(f'Skewness of "bmi": {df["bmi"].skew():.2f}')
print(f'Skewness of "charges": {df["charges"].skew():.2f}')
```

***We can figure out from the above visualizations and the skewness values that only the `BMI` column is normally distributed***

```python
# Plotting box-plots to find out the outliers
plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
sns.boxplot(df['age'])
plt.title('Box-plot of the "Ages"')

plt.subplot(1,3,2)
sns.boxplot(df['bmi'],color='red')
plt.title('Box-plot of the "BMIs"')

plt.subplot(1,3,3)
sns.boxplot(df['charges'],color='green')
plt.title('Box-plot of the "Charges')

plt.suptitle('Box-plot of each numerical column')

plt.tight_layout()
plt.show()
```

### From the above visualizations we can approximate the outliers:

- `Age` -> No outliers detected.
- `BMI` -> There are quite a few outliers present. The threshold value is around 47 (as can be seen in the above graph).
- `Charges` -> Lots of outliers are present in this column. The threshold value is around 35000 (as can be seen in the above graph).

## **Steps of doing Bivariate Analysis**

- Select 2 columns
- Understand type of relationship
    1. **Numerical - Numerical**<br>
        a. You can plot graphs like scatterplot(regression plots), 2D histplot, 2D KDEplots<br>
        b. Check correlation coefficent to check linear relationship
    2. **Numerical - Categorical** - create visualizations that compare the distribution of the numerical data across different categories of the categorical data.<br>
        a. You can plot graphs like barplot, boxplot, kdeplot violinplot even scatterplots<br>
    3. **Categorical - Categorical**<br>
        a. You can create cross-tabulations or contingency tables that show the distribution of values in one categorical column, grouped by the values in the other categorical column.<br>
        b. You can plots like heatmap, stacked barplots, treemaps
        
- Write your conclusions

#### **Numerical-Categorical Analysis**

```python
# Plotting the mean 'charge' across different regions

df.groupby('region')['charges'].mean().plot(kind='bar',color=['red','green','brown','blue'])
plt.ylabel('Avg. Charge')
plt.title('Avg. "Charge" Across Regions')

plt.tight_layout()
plt.show()
```

```python
# Plotting the mean 'charge' for smokers and non-smokers

df.groupby('smoker')['charges'].mean().plot(kind='bar',color=['red','blue'])
plt.ylabel('Avg. Charge')
plt.title('Avg. "Charge" for Smokers and Non-smokers')

plt.tight_layout()
plt.show()
```

```python
# Plotting the mean 'charge' for males and females

df.groupby('sex')['charges'].mean().plot(kind='bar',color=['green','brown'])
plt.ylabel('Avg. Charge')
plt.title('Avg. "Charge" for Males and Females')

plt.tight_layout()
plt.show()
```

```python
# Plotting the mean 'charge' for the number of children

df.groupby('children')['charges'].mean().plot(kind='bar')
plt.ylabel('Avg. Charge')
plt.title('Avg. "Charge" for No. of Chidren')

plt.tight_layout()
plt.show()
```

```python
# Plotting the mean 'age' across the regions

df.groupby('region')['age'].mean().plot(kind='bar')
plt.ylabel('Avg. Age')
plt.title('Avg. "Age" Across Regions')

plt.tight_layout()
plt.show()
```

```python
pd.crosstab(df['region'],df['smoker'],normalize='columns')*100
```

***This code creates a cross-tabulation table between the region and smoker columns in the dataframe df. The table is normalized by column, meaning that each column will sum to 100%. The output is multiplied by 100 to convert it to a percentage.***

```python
# Creating a heatmap of Smokers and Non-smokers across different regions

sns.heatmap(pd.crosstab(df['region'],df['smoker'],normalize='columns')*100)
plt.title('Smokers and Non-smokers Across Different Regions')
plt.tight_layout()
plt.show()
```

```python
pd.crosstab(df['region'],df['children'],normalize='columns')*100
```

***This code creates a cross-tabulation table between the region and children
